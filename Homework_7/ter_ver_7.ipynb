###1.	Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):   
``` zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]```,   
``` ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]```.        
Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без.

````python
import numpy as np
````

````python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
````

````python
# Вычисляем средние значения массивов
mean_zp = np.mean(zp)
mean_ks = np.mean(ks)

# Вычисляем отклонения от среднего
dev_zp = zp - mean_zp
dev_ks = ks - mean_ks

# Вычисляем коэффициенты a и b
a = np.sum(dev_zp * dev_ks) / np.sum(dev_zp**2)
b = mean_ks - a * mean_zp

# Выводим коэффициенты a и b
print("Коэффициенты линейной регрессии с учетом b: a = {}, b = {}".format(a, b))
````

Без использования intercept (то есть, без учета коэффициента b):

````python
# Вычисляем коэффициент a
a = np.sum(zp * ks) / np.sum(zp**2)

# Выводим коэффициент a
print("Коэффициент линейной регрессии без учета b: a = {}".format(a))
````

###2.	В каких случаях для вычисления доверительных интервалов и проверки статистических гипотез используется таблица значений функции Лапласа, а в каких - таблица критических точек распределения Стьюдента?

Таблица значений функции Лапласа используется для вычисления доверительных интервалов и проверки статистических гипотез в случае, когда известна или большая выборка из генеральной совокупности, и известна ее среднеквадратическая ошибка. Это означает, что можно использовать нормальное распределение для аппроксимации распределения выборочной статистики (например, среднего значения выборки или разности средних значений двух выборок), что позволяет использовать таблицу значений функции Лапласа для вычисления соответствующих квантилей.

С другой стороны, таблица критических точек распределения Стьюдента используется для вычисления доверительных интервалов и проверки статистических гипотез в случае, когда размер выборки мал или неизвестен и/или ошибка неизвестна. В этом случае используется распределение Стьюдента для аппроксимации распределения выборочной статистики, что требует использования таблицы критических точек распределения Стьюдента для определения соответствующих квантилей.

Таким образом, выбор таблицы зависит от того, какую информацию у нас есть о генеральной совокупности и о выборке. Если мы знаем среднеквадратическую ошибку генеральной совокупности и имеем большую выборку, мы можем использовать таблицу значений функции Лапласа. Если мы имеем маленькую выборку или не знаем ошибку генеральной совокупности, нам нужно использовать таблицу критических точек распределения Стьюдента.
